```python
import numpy as np
import matplotlib as mpl
```

```python
class LogisticRegression:
    def __init__(self, lr, nIter):
        self.lr = lr
        self.nIter = nIter
        self.w = None
        self.b = None
        self.lossHistory = []

    def _sigmoid(self, z):
        z = np.clip(z, -500, 500)
        return 1 / (1 + np.exp(-z))
    
    def fit(self, x, y):
        nSample, nFeature = x.shape
        self.w = np.zeros(nFeature)
        self.b = 0 
        for iter in range(self.nIter): 
            lin = np.dot(x, self.w) + self.b
            YPr = self._sigmoid(lin)
            epsilon = 1e-15
            YPr_clipped = np.clip(YPr, epsilon, 1- epsilon)
            loss = -np.mean(
                y * np.log(YPr_clipped) + (1-y) * np.log(1 - YPr_clipped)
            )
            self.loss_history.append(loss)
            dw = (1/nSample) * np.dot(x.T, (YPr - y))
            db = (1/nSample) * np.sum(YPr - y)
            self.w -= self.lr * dw
            self.b -= self.lr * db

            if iter % 100 == 0:
                print(f'итерация {iter:4d} loss {loss:.4f}')
    
    def predProb(self, x):
        lin = np.dot(x, self.w) + self.b
        return self._sigmoid(lin)
    
    def pred(self, x, threshold = 0.5):
        return (self.predProb(x) >= threshold).astype(int)
    
    def score(self, x, y):
        return np.mean(self.pred(x) == y)

```
