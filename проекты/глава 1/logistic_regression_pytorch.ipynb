```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#инициализация датасета
data = load_breast_cancer()
x, y = data.data, data.target

scaler = StandardScaler()
x = scaler.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
x_train = torch.tensor(x_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
x_test = torch.tensor(x_test, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)
```

```python
x_test.shape
```

```python
class LogRegression(nn.Module):
    def __init__(self, n_features):
        super().__init__()
        self.linear = nn.Linear(n_features, 1)

    def forward(self, x):
        return torch.sigmoid(self.linear(x))
    
```

```python
display(pd.DataFrame(x_train.numpy()))
```

```python
print(x_train.shape)
```

```python
model = LogRegression(n_features=x_train.shape[1])

crit = nn.BCELoss()
opt = optim.SGD(model.parameters(), lr = 0.01)

n_epochs = 10000
for epoch in range(n_epochs):
    y_pred = model(x_train)
    loss = crit(y_pred, y_train)
    opt.zero_grad()
    loss.backward()
    opt.step()
    if epoch % 100 == 0:
        print(f'эпоха {epoch:4d}, лосс {loss.item():.4f}')
    
```

```python
with torch.no_grad():
    y_pred_test = (model(x_test) >= 0.5).float()

y_pred_test.shape
y_test.shape
```

```python
with torch.no_grad():
    acc = (y_pred_test == y_test).float().mean()
    print(acc)
```
